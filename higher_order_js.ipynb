```python
%%writefile code.js
// Part A
const data = [5, 12, 3, 20, 11, 9, 30];
let loopData = []
// loop version
for (let i = 0; i < data.length; i++) {
  if (data[i] > 10) {
    loopData.push(data[i] * 2);
  }
}

// pipeline version
const pipelineData = data
  .filter(val => val > 10)
  .map(val => val * 2);

console.log("Part A Loop:", loopData);
console.log("Part A Pipeline:", pipelineData);


// Part B
const numbers = [1, 2, 3, 4, 5, 6];

const sumOfDoubledEvens = numbers
  .filter(n => n % 2 === 0) // keep evens
  .map(n => n * 2) // double them
  .reduce((acc, val) => acc + val, 0); // add them all up

console.log("Part B:", sumOfDoubledEvens);


// Part C
console.log("\nPart C:")
const users = [
  { id: 1, name: "Parker", age: 21 },
  { id: 2, name: "Faye", age: 20 },
  { id: 3, name: "Donny", age: 20 }];

console.log("Any under 21?", users.some(u => u.age < 21));
console.log("Find ID 3:", users.find(u => u.id === 3));
users.forEach(u => console.log(u.name));


// Part D
function repeat(n, action) {
  for (let i = 0; i < n; i++) {
        action(i);
  }
}

console.log("\nPart D Tests:");
repeat(3, console.log);
repeat(4, i => console.log("Step " + i));
```
